# Polynom data data generation

This notebook generates very basic datasets of sinus time-series, that get modulated by polynoms of different degrees. The datasets are saved as `.csv`.

```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

SEQ_LEN = 10000
```

```python
def fct1(x, a, b):
    y = a * x + b
    return y

def fct2(x, a, b, c):
    y = a * (x**2) + b * x + c
    return y

def fct3(x, a, b, c, d):
    y = a * (x**3) + b * (x**2) + c * x + d
    return y

def base_fct(t):
    x = np.sin(np.pi * t)
    #x = signal.square(x)
    return x
```

## Complexity first order

sin fct with polynomial transformation of first degree
param: a and b 

- network input: sin + a param + b param 
- network output: transformed sin

```python
# ds modulation 
a, b = 8, 4

def create_1o_fct(a, b, seq_len):
    t = np.linspace(0, 10, seq_len, endpoint=False)
    x = base_fct(t)
    y = fct1(x, a, b)
    param_a = [a] * seq_len
    param_b = [b] * seq_len
    return x, param_a, param_b, y

x, param_a, param_b, y = create_1o_fct(a, b, SEQ_LEN)

plt.figure(figsize=(20,5))
plt.subplot(3, 1, 1)
plt.plot(x, label="input")
plt.legend(loc="upper right")
plt.title("input")

plt.subplot(3, 1, 2)
plt.plot(param_a, label="param_a")
plt.plot(param_b, label="param_b")
plt.legend(loc="upper right")
plt.title("parameters")

plt.subplot(3, 1, 3)
plt.plot(y, label="output")
plt.legend(loc="upper right")
plt.title("output")
```

```python
x, y, param_a, param_b = [], [], [], []

for i in range(1, 4):
    for j in range(1, 4):
        x_, param_a_, param_b_, y_ = create_1o_fct(i * 4, j * 2, SEQ_LEN)
        x = np.append(x, x_)
        param_a = np.append(param_a, param_a_)
        param_b = np.append(param_b, param_b_)
        y = np.append(y, y_)
    
    
plt.figure(figsize=(20,5))
plt.subplot(3, 1, 1)
plt.plot(x, label="input")
plt.legend(loc="upper right")
plt.title("input")

plt.subplot(3, 1, 2)
plt.plot(param_a, label="param_a")
plt.plot(param_b, label="param_b")
plt.legend(loc="upper right")
plt.title("parameters")

plt.subplot(3, 1, 3)
plt.plot(y, label="output")
plt.legend(loc="upper right")
plt.title("output")
```

```python
# export to csv
df = pd.DataFrame()

df["x"] = x
df["param_a"] = param_a
df["param_b"] = param_b
df["y"] = y

df.to_csv("ds1o.csv")
```

```python
x, y, param_a, param_b, param_c = [], [], [], [], []

for i in range(1, 4):
    for j in range(1, 4):
        x_, param_a_, param_b_, param_c_, y_ = create_2o_fct(i * 4, j * 2, 4, SEQ_LEN)
        x = np.append(x, x_)
        param_a = np.append(param_a, param_a_)
        param_b = np.append(param_b, param_b_)
        param_c = np.append(param_c, param_c_)
        y = np.append(y, y_)
    
    
plt.figure(figsize=(20,5))
plt.subplot(3, 1, 1)
plt.plot(x, label="input")
plt.legend(loc="upper right")
plt.title("input")

plt.subplot(3, 1, 2)
plt.plot(param_a, label="param_a")
plt.plot(param_b, label="param_b")
plt.plot(param_c, label="param_c")
plt.legend(loc="upper right")
plt.title("parameters")

plt.subplot(3, 1, 3)
plt.plot(y, label="output")
plt.legend(loc="upper right")
plt.title("output")
```
